```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define computational domain size
Nx, Ny, Nz = 100, 100, 100  # Grid size
resolution = 0.1  # Grid resolution (points per unit length)

# Create a 3D computational cell
cell_size = mp.Vector3(Nx, Ny, Nz)

# Define a Gaussian point source at the center
wavelength = 1.0  # Central wavelength
freq = 1 / wavelength  # Frequency
source = mp.Source(
    src=mp.GaussianSource(frequency=freq, fwidth=0.2 * freq),
    component=mp.Ez,  # Choosing Ez as the polarization component
    center=mp.Vector3(0, 0, 0)  # Positioned at the center
)
```

```python
# Set up the simulation with vacuum and a single source
sim = mp.Simulation(
    cell_size=cell_size,
    resolution=resolution,
    sources=[source],
    boundary_layers=[mp.PML(1.0)],  # Absorbing boundaries to avoid reflections
)
```

```python
# Run the simulation
sim.run(until=100)
```

```python
# Visualizing the Ez field in the xy-plane at z=0
eps_data = sim.get_array(center=mp.Vector3(), size=mp.Vector3(Nx, Ny, 0), component=mp.Ez)
plt.imshow(eps_data.transpose(), cmap='RdBu', interpolation='spline36')
plt.colorbar()
plt.title("Ez field at z=0")
plt.show()
```
